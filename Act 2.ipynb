```python
import matplotlib.pyplot as plt
import random
import math

# Parámetros
num_circles = 20
circle_radius = 0.05
canvas_size = (1, 1)  # Tamaño del área (1x1)

# Función para verificar colisiones
def check_collision(new_circle, existing_circles, radius):
    for circle in existing_circles:
        distance = math.sqrt((new_circle[0] - circle[0])**2 + (new_circle[1] - circle[1])**2)
        if distance < 2 * radius:
            return True
    return False

# Generar círculos
circles = []
while len(circles) < num_circles:
    x = random.uniform(circle_radius, canvas_size[0] - circle_radius)
    y = random.uniform(circle_radius, canvas_size[1] - circle_radius)
    if not check_collision((x, y), circles, circle_radius):
        circles.append((x, y))

# Dibujar los círculos
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.set_xlim(0, canvas_size[0])
ax.set_ylim(0, canvas_size[1])

for circle in circles:
    circle_patch = plt.Circle(circle, circle_radius, color='blue')
    ax.add_patch(circle_patch)

plt.title("Círculos sin colisiones")
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Parámetros
num_circles = 20
circle_radius = 0.05
canvas_size = (1, 1)  # Tamaño del área (1x1)

# Calcular posiciones equidistantes centradas
def generate_equidistant_positions(num_circles, radius, canvas_size):
    positions = []
    grid_size = int(np.ceil(np.sqrt(num_circles)))
    step = min(canvas_size[0], canvas_size[1]) / grid_size
    start_offset_x = (canvas_size[0] - (grid_size - 1) * step) / 2
    start_offset_y = (canvas_size[1] - (grid_size - 1) * step) / 2

    for i in range(grid_size):
        for j in range(grid_size):
            if len(positions) >= num_circles:
                break
            x = start_offset_x + i * step
            y = start_offset_y + j * step
            if x + radius <= canvas_size[0] and y + radius <= canvas_size[1]:
                positions.append((x, y))
    return positions

circles = generate_equidistant_positions(num_circles, circle_radius, canvas_size)

# Dibujar los círculos
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.set_xlim(0, canvas_size[0])
ax.set_ylim(0, canvas_size[1])

for circle in circles:
    circle_patch = plt.Circle(circle, circle_radius, color='blue')
    ax.add_patch(circle_patch)

plt.title("Círculos espaciados equitativamente y centrados")
plt.show()

```
